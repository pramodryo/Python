# Q4 Find out the correlation between CM Satisfaction and PM Satisfaction Create the heat map for the same

```python
# type 1 : for calculate correlation 

# find the correlation between string values
data["satisfation_cm"] = data["satisfation_cm"].astype('category').cat.codes   
data["satisfation_PM"] = data["satisfation_PM"].astype('category').cat.codes

# cat.codes : convert your category from a string representation into an ineger representaion


correlation = data.corr()
correlation
```

```python
plt.figure(figsize= (10,6))
dataplot = sns.heatmap(correlation, cmap = "cubehelix_r", annot = True)

plt.show()
```

```python
# type 2 : for calculate correlation 

# By using corr() function we can get the correlation between two columns in the dataframe

corr = data["satisfation_cm"].corr(data["satisfation_PM"])
corr
```

# Q8 Create a pie chart showcasing the number of votes obtained by each party in vs_17 and ls_19									
									


```python
data.head(2)
```

```python
is_19 = data["ls_19"].value_counts().reset_index()
is_19
```

```python
type(is_19["ls_19"])
```

```python
plt.figure(figsize=(15,10))
plt.pie(is_19['ls_19'], labels = is_19['index'], startangle=90, autopct ='%2.1f%%')
plt.legend(loc='best')


# sns.set(rc = {"figure.figsize" : (10,10)}) # for figure size

plt.show()
```

```python
vs_17 = data["vs_17"].value_counts().reset_index()
vs_17
```

```python
plt.pie(vs_17['vs_17'], labels = vs_17['index'], startangle=90, autopct ='%2.1f%%')
plt.legend(loc='best')


sns.set(rc = {"figure.figsize" : (15,10)}) # for figure size

plt.show()
```

```python
temp_df = pd.merge(is_19,vs_17, on= "index",how = "right")   
# type of how ("left","right","outer","inner","cross" and default "inner")

temp_df["total_votes"] = temp_df[["ls_19","vs_17"]].sum(axis= 1)
temp_df
```

```python
plt.pie(temp_df['total_votes'], labels = temp_df['index'], startangle=90, autopct ='%2.1f%%')
plt.legend(loc='best')


sns.set(rc = {"figure.figsize" : (10,10)}) # for figure size

plt.show()
```
