#  Subplots

### a. Display Multiple Plots

```python
# draw 2 plots

# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

#plotting subplot-1
plt.subplot(1,2,1) # the fig has 1 raw, 2 column and this is the first plot
plt.plot(x1,y1)

# plotting subplot-2
plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.show()
```

```python
# Draw 2 plots

# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

# plotting subplot-1
plt.subplot(2,1,1) # the figure has 2 row, 1 columns and this is the first plot
plt.plot(x1,y1)

# plotting subplot-2
plt.subplot(2,1,2)
plt.plot(x2,y2)

plt.show()
```

```python
# Draw 6-plots
# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

# plot-3 parameters
x3 = np.array([0,1,2,3])
y3 = np.array([3,8,1,10])

# plot-4 parameters
x4 = np.array([0,1,2,3])
y4 = np.array([10,20,30,40])

# plot-5 parameters
x5 = np.array([0,1,2,3])
y5 = np.array([3,8,1,10])

# plot-6 parameters
x6 = np.array([0,1,2,3])
y6 = np.array([10,20,30,40])

plt.subplot(2,3,1)
plt.plot(x1,y1)

plt.subplot(2,3,2)
plt.plot(x2,y2)

plt.subplot(2,3,3)
plt.plot(x3,y3)

plt.subplot(2,3,4)
plt.plot(x4,y4)

plt.subplot(2,3,5)
plt.plot(x5,y5)

plt.subplot(2,3,6)
plt.plot(x6,y6)

plt.show()
```

```python
# Draw 6-plots : correcting the distance between the plots
# plot-1 parameters
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])

# plot-2 parameters
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

# plot-3 parameters
x3 = np.array([0,1,2,3])
y3 = np.array([3,8,1,10])

# plot-4 parameters
x4 = np.array([0,1,2,3])
y4 = np.array([10,20,30,40])

# plot-5 parameters
x5 = np.array([0,1,2,3])
y5 = np.array([3,8,1,10])

# plot-6 parameters
x6 = np.array([0,1,2,3])
y6 = np.array([10,20,30,40])

plt.subplot(2,3,1)
plt.plot(x1,y1)
plt.title('x1 vs y1')

plt.subplot(2,3,2)
plt.plot(x2,y2)
plt.title('x2 vs y2')

plt.subplot(2,3,3)
plt.plot(x3,y3)
plt.title('x3 vs y3')

plt.subplot(2,3,4)
plt.plot(x4,y4)
plt.title('x4 vs y4')

plt.subplot(2,3,5)
plt.plot(x5,y5)
plt.title('x5 vs y5')

plt.subplot(2,3,6)
plt.plot(x6,y6)
plt.title('x6 vs y6')

plt.suptitle("Sales Data")  #super title

plt.tight_layout()
```

#  Matplotlib Bars

```python
# vertical bar plots
x=np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.bar(x,y)
plt.show()
```

```python
# horizantal bar plots
x=np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.barh(x,y) # horizantal bar charts
plt.show()
```

```python
# vertical bar plots, coloring bar

x=np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.barh(x,y, color="red")
plt.show()
```

```python
# Bar width : it will work only in vertical chart

x=np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.bar(x,y, color="red", width= 0.2)
plt.show()
```

```python
# bar height works in horizantle charts

x=np.array(['A','B','C','D','E'])
y = np.array([4,9,2,11,16])

plt.barh(x,y, color="red", height = 0.1)
plt.show()
```

# Histogram

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
# create a histogram with 15 buckets/bins

x = np.random.normal(170,10,250)
plt.hist(x,bins=15)
plt.show()
```

```python
# create a histogram with density

x =np.random.normal(170,10,250)
plt.hist(x,bins=15, density = True)
plt.show()


# formula for density = counts/(sum(counts) * np.diff(bins))
```

```python
# plotting two histogram togather: side by side plots

y = np.random.normal(180,20,150)
plt.hist([x,y],stacked=False)

plt.show()
```

```python
# plotting two histogram togather: stacked plot: one plot on top of other plot

y = np.random.normal(180,20,150)
plt.hist([x,y],stacked=True)

plt.show()
```

```python
#Adding the legend
# Beautiful pie chart: add shadow and colors
# exploding the pices
# insert label to this pie chart


y = np.array([35,25,25,15])
my_labels = ['English','Math','Science','Hindi']
my_explode = [0,0,0.1,0.3]
colors = ['orange','black','magenta','cyan']
plt.pie(y, labels = my_labels, startangle=90, explode=my_explode,colors=colors, shadow=True)
plt.legend(loc='best')
plt.show()
```

```python
# add header to the legend
#Adding the legend
# Beautiful pie chart: add shadow and colors
# exploding the pices
# insert label to this pie chart


y = np.array([35,25,25,15])
my_labels = ['English','Math','Science','Hindi']
my_explode = [0,0,0.1,0.3]
colors = ['orange','black','magenta','cyan']
plt.pie(y, labels = my_labels, startangle=90, explode=my_explode,colors=colors, shadow=True, autopct ='%2.1f%%')
plt.legend(title='Marks', loc='best')
plt.show()
```
